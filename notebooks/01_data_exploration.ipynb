```python
# 01_data_exploration.ipynb
# Exploratory Data Analysis for Customer Support Ticket Dataset

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
df = pd.read_csv(r'C:\Users\ibrahim.fadhili\OneDrive - Agile Business Solutions\Desktop\Customer Support\data\customer_support_tickets.csv')
```

```python
# Basic info
print("Dataset Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nMissing Values:\n", df.isnull().sum())
```

```python
# Summary statistics
print("\nSummary Statistics:\n", df.describe())
```

```python
# Visualizations
# 1. Ticket Type distribution
plt.figure(figsize=(8, 6))
df['Ticket Type'].value_counts().plot(kind='bar', title='Ticket Type Distribution')
plt.xlabel('Ticket Type')
plt.ylabel('Count')
plt.show()
```

```python
# 2. Resolution Time by Priority
plt.figure(figsize=(8, 6))
sns.boxplot(x='Ticket Priority', y='Time to Resolution', data=df)
plt.title('Resolution Time by Ticket Priority')
plt.show()
```

```python
# 4. Ticket Channel distribution
plt.figure(figsize=(8, 6))
df['Ticket Channel'].value_counts().plot(kind='pie', title='Ticket Channel Distribution')
plt.show()
```
